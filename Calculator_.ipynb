```python
import tkinter as tk

root = tk.Tk()
root.title("Hat's Calculator")

root.configure(bg="#333333")  

display_frame = tk.Frame(root, bd=10, relief="sunken", bg="#444444")
display_frame.pack(expand=True, fill="both")

display = tk.Entry(display_frame, font=("Arial", 24), bg="#222222", fg="white", bd=10, relief="flat", justify="right")
display.pack(expand=True, fill="both")

current_expression = ""

def button_click(item):
    global current_expression
    current_expression += str(item)
    display.delete(0, tk.END)
    display.insert(tk.END, current_expression)

def clear_display():
    global current_expression
    current_expression = ""
    display.delete(0, tk.END)

def evaluate_expression():
    global current_expression
    try:
        result = str(eval(current_expression))  
        display.delete(0, tk.END)
        display.insert(tk.END, result)
        current_expression = result
    except:
        display.delete(0, tk.END)
        display.insert(tk.END, "Error")
        current_expression = ""


button_frame = tk.Frame(root, bg="#333333")
button_frame.pack(expand=True, fill="both")

buttons = [
    ('7', '8', '9', '/'),
    ('4', '5', '6', '*'),
    ('1', '2', '3', '-'),
    ('C', '0', '=', '+')
]

button_colors = {
    '/': "#FF6666",
    '*': "#FF6666",
    '-': "#FF6666",
    '+': "#FF6666",
    'C': "#FF9933",
    '=': "#66CC66",
    'default': "#CCCCCC"
}

for row in buttons:
    row_frame = tk.Frame(button_frame, bg="#333333")
    row_frame.pack(expand=True, fill="both")
    for char in row:
        button_color = button_colors.get(char, button_colors['default'])
        button = tk.Button(
            row_frame,
            text=char,
            font=("Arial", 18),
            fg="black",
            bg=button_color,
            bd=0,
            relief="flat",
            command=lambda x=char: button_click(x) if x not in ['=', 'C'] else evaluate_expression() if x == '=' else clear_display()
        )
        button.pack(side="left", expand=True, fill="both")

# Run the application
root.mainloop()
```
